```python
import matplotlib.pyplot as plt
import numpy as np
import os
import cv2
import random
#import tensorflow as tf

dir='F:/Cats and Dogs/PetImages' #path where image is stored
categories=["Dog","Cat"]
for category in categories:
    path=os.path.join(dir,category)
    for img in os.listdir(path):
        img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
        break
    break

IMG_SIZE=50
new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
plt.imshow(new_array,cmap='gray')
plt.show()
```

```python
training_data=[]
def create_training_data():
    for category in categories:
        path=os.path.join(dir,category)
        class_num=categories.index(category)
        for img in (os.listdir(path)):
            try:
                img_array=cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,class_num])
                dr=path.split('/')[-1]
              # print("evaluating image of ",category," image no=",img)
            except Exception as e:
                pass
create_training_data()
print(len(training_data))

random.shuffle(training_data)
# Make Model
X=[]
y=[]
for features , label in training_data:
    X.append(features)
    y.append(label)
X=np.array(X).reshape(-1,IMG_SIZE,IMG_SIZE,1)
```

```python
import pickle
import time
import keras
from keras.models import load_model

#from tensorflow.keras.callbacks import TensorBoard

# pickle_out=open("X.pickle","wb")
# pickle.dump(X,pickle_out)
# pickle_out.close()
# pickle_out=open("y.pickle","wb")
# pickle.dump(y,pickle_out)
# pickle_out.close()

# NAME="Cats-Dogs-cnn-64x2-{}".format(int(time.time()))
# pickle_in=open("X.pickle","rb")
# X=pickle.load(pickle_in)
# pickle_in=open("y.pickle","rb")
# y=pickle.load(pickle_in)

X=X/255.0
#tensorboard=TensorBoard(log_dir='logs/{}'.format(NAME))

from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dropout,Dense,Activation,Flatten, Conv2D, MaxPooling2D

model=Sequential()
model.add(Conv2D(256,(3,3),activation='relu',input_shape=X.shape[1:]))
#model.add(Dropout(0.25))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(256,(3,3),activation='relu'))
model.add(Dropout(0.2))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',metrics=['accuracy'])

X_train=X[0:23000,:]
y_train=y[0:23000]

#plt.imshow(X[1],cmap='gray')

X[0]
X_valid=X[23000:,:]

y_valid=y[23000:]
# len(y_test)
# X_test.shape
model.summary()
```

```python
hist=model.fit(X_train,y_train,batch_size=128,epochs=10,validation_data=(X_valid,y_valid))
#plt.imshow(X.reshape(-1,50,50,1))
```
